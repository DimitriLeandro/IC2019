```python
def listarDispositivos():
    p = pyaudio.PyAudio()
    info = p.get_host_api_info_by_index(0)
    numdevices = info.get('deviceCount')

    for i in range(0, numdevices):
        if (p.get_device_info_by_host_api_device_index(0, i).get('maxInputChannels')) > 0:
            print("Input Device id ", i, " - ", p.get_device_info_by_host_api_device_index(0, i).get('name'))
```

### Definição de funções

```python
def main_1(idDispositivoGravacao, tempoJanela, freqAmostragem, profundidadeBytes, qtdCanais):
    
    # DEFININDO O TAMANHO DA JANELA EM AMOSTRAS
    tamanhoJanela = int(tempoJanela * freqAmostragem)
    
    # LOOP
    i_1 = 0
    while i_1 < 10:       

        # VERBOSE
        i_1 += 1
        stringID_1 = "(main_1)(id:" + str(i_1) + ")"

        # INSTANCIANDO UM OBJ PY AUDIO E MANDANDO OS PARAMETROS INICIAIS
        objPyAudio_1 = pyaudio.PyAudio()

        # TENTANDO ABRIR O STREAM ATE IR
        stream_1 = None
        while stream_1 == None:
            print(stringID_1, "Tentando abrir uma stream")
            try:
                stream_1 = objPyAudio_1.open(
                    input_device_index = idDispositivoGravacao,
                    rate               = freqAmostragem,
                    format             = objPyAudio_1.get_format_from_width(profundidadeBytes),
                    channels           = qtdCanais,
                    input              = True
                )
            except:
                pass

        # VERBOSE
        inicioGravacao = time()
        print(stringID_1, "Iniciando gravação. Timestamp:", inicioGravacao)

        # GRAVANDO A JANELA
        janelaBinaria_1 = stream_1.read(tamanhoJanela)

        # MATANDO OS OBJETOS PRA LIMPAR MEMORIA
        stream_1.stop_stream()
        stream_1.close()
        objPyAudio_1.terminate()
        del stream_1
        del objPyAudio_1

        # INICIANDO A MEDICAO DE TEMPO DE PROCESSAMENTO
        inicioProcessamento = time()
        print(stringID_1, "Iniciando processamento. Timestamp:", inicioProcessamento)

        # CONVERTO A JANELA PRA INT 16
        janelaInt16_1 = np.fromstring(janelaBinaria_1, dtype=np.int16)

        # COMO TEM OS 4 MICS NA JANELA, VOU DAR UM RESHAPE PRA CADA MIC FICAR EM UMA LINHA
        janelaInt16_1 = janelaInt16_1.reshape((tamanhoJanela, qtdCanais)).T    

        # DELAY AND SUM
        sinalDaS_1 = DaS(janelaInt16_1)

        # EXTRACAO DE FEATURES
        features_1 = ExtrairFeatures(sinalDaS_1, freqAmostragem)

        # NORMALIZANDO AS FEATURES
        features_1 = objNormalizador.transform([features_1])[0]

        # CLASSIFICACAO
        predicao_1 = objClassificador.predict([features_1])[0]

        # FINALIZANDO A MEDICAO DE TEMPO
        fimProcessamento = time()
        print(stringID_1, "Processamento finalizado. Timestamp:", fimProcessamento)
        
        # ADICIONANDO NO DICIONARIO GLOBAL PARA PLOTAR DEPOIS
        dictGlobal[stringID_1] = {
            "inicioGravacao": inicioGravacao,
            "inicioProcessamento": inicioProcessamento,
            "fimProcessamento": fimProcessamento
        }
```

### Printando

```python
def corAleatoria():
    return "#" + str(randint(10,99)) + str(randint(10,99)) + str(randint(10,99))
```

```python
#DEFININDO ALGUNS PARÂMETROS DO GRÁFICO
%matplotlib inline
%config InlineBackend.figure_format = 'svg'
params = {
    'figure.figsize': [11, 5.5], 
    'axes.labelsize': 14,
    'axes.titlesize':18, 
    'font.size': 18,
    'legend.fontsize': 12, 
    'xtick.labelsize': 10, 
    'ytick.labelsize': 10,
    'axes.axisbelow': True
}
plt.rcParams.update(params)
```

```python
yTicks = []

for i, chave in enumerate(dictGlobal):
    
    # AJUSTANDO O EIXO X DA ITERACAO ATUAL
    inicioGravacao      = dictGlobal[chave]["inicioGravacao"] - tempoReferencia
    fimGravacao         = inicioGravacao + tempoJanela
    inicioProcessamento = dictGlobal[chave]["inicioProcessamento"] - tempoReferencia
    fimProcessamento    = dictGlobal[chave]["fimProcessamento"] - tempoReferencia
    
    # DEFININDO A COR
    corAtual = corAleatoria()
    
    # COLOCANDO A ITERACAO ATUAL NO YTICKS
    yTicks.append(chave)
    
    # PLOTANDO
    plt.plot([inicioGravacao, fimGravacao], [i, i], "-", color=corAtual)
    plt.plot([inicioProcessamento, fimProcessamento], [i, i], "--", color=corAtual)

# CRIANDO A LEGENDA
plt.plot([fimProcessamento,fimProcessamento],[i,i],"-",label="Gravação", color="black")
plt.plot([fimProcessamento,fimProcessamento],[i,i],"--",label="Processamento", color="black")

# AJUSTANDO MAIS PARAMETROS
plt.title("Diagrama de Gantt - Raspberry")
plt.ylabel("Thread")
plt.xlabel("Tempo (s)")
plt.legend()
plt.xticks(ticks=np.arange(0, fimProcessamento+tempoJanela, tempoJanela), rotation=45)
plt.yticks(ticks=np.arange(0, i+1), labels=yTicks)
plt.gca().invert_yaxis()
plt.grid(alpha=0.4)
plt.show()
```
