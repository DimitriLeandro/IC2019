# Extraindo features para diversos tamanhos de janelamento

```python
import os
import numpy as np
import librosa
import sys
sys.path.append('/home/dimi/Programming/IC2019/ML/python')
from extrairFeaturesClass import ExtrairFeatures
```

### Verificando o maior janelamento possível

```python
dirOrigem    = '/home/dimi/Downloads/datasets/SESA_v2_16kHz_16bits/original/'
menorDuracao = 100
for arquivoAtual in os.listdir(dirOrigem):
    sinalAtual, freqAmostragem = librosa.load(dirOrigem+arquivoAtual, sr=None, mono=True)
    duracaoArquivoAtual = librosa.core.get_duration(y=sinalAtual, sr=freqAmostragem)
    if duracaoArquivoAtual < menorDuracao:
        menorDuracao = duracaoArquivoAtual
print(menorDuracao)
```

### Vamo que vamo

```python
# DEFININDO PARAMETROS INICIAIS
dirOrigem      = '/home/dimi/Downloads/datasets/SESA_v2_16kHz_16bits/original/'
dirDestino     = '/home/dimi/Downloads/datasets/SESA_v2_16kHz_16bits/'
freqAmostragem = 16000
escalonamento  = False
janelamentos   = np.arange(0.025, menorDuracao, 0.025)

# PARA CADA JANELAMENTO
for janelamentoAtual in np.flip(janelamentos):
    
    # DEFININDO O TAMANHO DO JANELAMENTO EM AMOSTRAS
    frameLength    = int(janelamentoAtual * freqAmostragem)
    overlapLength  = int(frameLength / 2)

    # EXTRAINDO
    ExtrairFeatures(dirOrigem, freqAmostragem, frameLength, overlapLength, escalonamento)
    
    # TIRANDO O CSV DA PASTA DE ORIGEM PRA PROXIMA ITERACAO FUNCIONAR (SO PODE TER WAV NA PASTA)
    arquivoCSVGerado = [arquivoAtual for arquivoAtual in os.listdir(dirOrigem) if arquivoAtual[-4:] == ".csv"][0]
    os.rename(dirOrigem+arquivoCSVGerado, dirDestino+"SESA_v2_16kHz_16bits_"+str(janelamentoAtual*100).split["."][0]+"ms_58features_desescalonado_remocaoSilencio.csv")
```
