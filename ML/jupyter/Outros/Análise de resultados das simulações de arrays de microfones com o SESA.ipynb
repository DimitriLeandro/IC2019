## Começando

```python
rangeSNRs = list(np.arange(-10, 31))
```

```python
# ORIGEM DE ONDE VAO ESTAR AS PASTAS COM OS SNRS CHEIOS DE RELATORIOS DE CLASSIFICADORES
diretorioOrigem  = "/home/dimi/Programming/IC2019/ML/relatorios_classificacao/SESA_Normalizado/data_augmentation/ruido_branco/com_gridsearch/"

# PARA CADA PASTA DE SNR
for SNRAtual in rangeSNRs:
    
    # PEGANDO OS NOMES DE TODOS OS JSONS DENTRO DESSA PASTA DE SNR ATUAL
    diretorioSNRAtual = diretorioOrigem + "SNR_" + str(SNRAtual) + "dB/"
    arrayArquivosJSONSNRAtual = [arquivo for arquivo in os.listdir(diretorioSNRAtual) if arquivo.endswith(".json")]
    
    # PARA CADA JSON
    for nomeJSONAtual in arrayArquivosJSONSNRAtual:
        
        # EU ABRO O JSON
        with open(diretorioSNRAtual + nomeJSONAtual) as arquivoJSON:
            dadosJSON = json.load(arquivoJSON)

            # PEGO A ACURACIA ATUAL
            acuraciaAtual = dadosJSON["accuracy"]

            # VERIFICO ONDE EU TENHO QUE COLOCAR NO DICIONARIO
            keySNR         = "SNR_" + str(SNRAtual) + "dB"
            classificador  = dadosJSON["classificador"]
            tipoBeamfoming = nomeJSONAtual.split("_")[2]
            
            # E COLOCO NO LUGAR CERTO 
            dicionarioDados[keySNR][classificador][tipoBeamfoming].append(acuraciaAtual)
```
