# Desempenho do Classificador em Função do Tamanho do Janelamento do Dataset

Esse jupyter deverá ser rodado na servidora. Lá, temos CSVs do SESA v.2 16 kHz 16 bits com diferentes janelamentos utilizados.

A ideia desse jupyter é verificar qual é o melhor tamanho de janela para a acurácia do modelo. Já sabemos que, para rodar na Raspberry, o classificador será o SGD, que vamos usar somente 58 features (sem PCA), e que o dataset deve ter profundidade de 16 bits. 

Os resultados desse jupyter deverão ser agragados aos resultados do experimento que mediu o tempo de processamento em função do janelamento na Raspberry. Com isso, vamos descobrir qual é o melhor tamanho de janela para usar no sistema em tempo real que vai rodar na Raspberry.

### Importações

```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import time
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score
from sklearn.linear_model import SGDClassifier
```

```python
# DEFININDO ALGUNS PARÂMETROS PARA PLOTAR GRÁFICOS
%matplotlib inline
%config InlineBackend.figure_format = 'svg'
params = {'figure.figsize': [11, 5.5], 
          'axes.labelsize': 12,
          'axes.titlesize':12, 
          'font.size': 11,
          'legend.fontsize': 11, 
          'xtick.labelsize': 10, 
          'ytick.labelsize': 10
         }
plt.rcParams.update(params)
```

### Parâmetros Iniciais

```python
# DIRETORIO QUE CONTEM OS CSVS DO DATASET - UM PARA CADA TAMANHO DE JANELAMENTO
diretorioCSVs = "/home/dimi/Downloads/datasets/SESA_v2_16kHz_16bits/"

# QUANTIDADE DE FOLDS QUE O DATASET ESTA SEGMENTADO
qtdPastas = 5

# DIRETORIO PARA SALVAR OS RESULTADOS
caminhoSalvar = "/home/dimi/Programming/IC2019/Raspberry/Resultados/"
```

### Definição de Funções

```python
def normalizarFeatures(data):
    objNormalizador = StandardScaler()
    objNormalizador.fit(data)
    data = objNormalizador.transform(data)
    return data
```

```python
def ajustarDataset(caminhoCSV, pasta):
    
    # ABRINDO
    dataframe = pd.read_csv(caminhoCSV)
    
    # SEPARANDO DATA, TARGET E O NOME DOS ARQUIVOS
    data     = dataframe.iloc[:, 1:-1].to_numpy()
    target   = dataframe.iloc[:, -1].to_numpy()
    arquivos = dataframe.iloc[:, 0].to_numpy()
    
    # LIMPANDO A MEMORIA
    del dataframe
    
    # NORMALIZANDO AS FEATURES
    data = normalizarFeatures(data)
    
    # SEPARANTE TREINO E TESTE PELO NOME DO ARQUIVO (O ULTIMO NUMERO NO NOME E A PASTA (casual_000_1.wav)
    xTrain, xTest, yTrain, yTest = [], [], [], []
    for xAtual, yAtual, arquivoAtual in zip(data, target, arquivos):
        
        # SE FOR DA PASTA ESCOLHIDA VAI PRO TESTE
        if int(arquivoAtual.split("_")[2][0]) == pasta:
            xTest.append(xAtual)
            yTest.append(yAtual)
        
        # SENAO VAI PRO TREINAMENTO
        else:
            xTrain.append(xAtual)
            yTrain.append(yAtual)  
    
    # PASSANDO TUDO PRA NUMPY
    xTrain = np.array(xTrain)
    xTest  = np.array(xTest)
    yTrain = np.array(yTrain)
    yTest  = np.array(yTest)
            
    # VERIFICANDO A DIMENSIONALIDADE
    assert xTrain.shape[0] + xTest.shape[0] == data.shape[0], "Dimensionalidade dos Xs não bate."
    assert yTrain.shape[0] + yTest.shape[0] == target.shape[0], "Dimensionalidade dos Ys não bate."
    assert xTrain.shape[1] == xTest.shape[1] and xTest.shape[1] == data.shape[1], "Número de features não bate."
    
    # VAMO QUE VAMO
    return xTrain, xTest, yTrain, yTest
```

```python
def treinarEClassificar(xTrain, xTest, yTrain, yTest):
    
    # INSTANCIANDO O CLASSIFICADOR
    objClassificador = SGDClassifier(alpha=0.0001, average=False, class_weight=None, early_stopping=False, epsilon=0.1, eta0=0.0, fit_intercept=True, l1_ratio=0.15, learning_rate='optimal', loss='log', max_iter=1000, n_iter_no_change=5, n_jobs=-1, penalty='l2', power_t=0.5, random_state=None, shuffle=True, tol=0.0001, validation_fraction=0.1, verbose=0, warm_start=False)
    
    # TREINANDO
    objClassificador.fit(xTrain, yTrain)
    
    # PREDIZENDO
    yPred = objClassificador.predict(xTest)
    
    # RETORNANDO A ACURACIA
    return accuracy_score(yTest, yPred)
```

```python
def iterarKFold(caminhoCSV, qtdPastas):
    
    acuracias = []
    
    # PARA CADA PASTA DO K FOLD
    for pastaAtual in range(1, qtdPastas+1):
        
        # OBTENHO A SEPARACAO DE TREINENTO E TESTE
        xTrain, xTest, yTrain, yTest = ajustarDataset(caminhoCSV, pastaAtual)
        
        # TREINO, CLASSIFICO E OBTENHO A ACURACIA
        acuracias.append(treinarEClassificar(xTrain, xTest, yTrain, yTest))
        
    # DEVOLVO AS METRICAS
    return np.mean(acuracias), np.std(acuracias)
```

### Rodando o algoritmo

```python
# OBTENDO UM ARRAY COM OS ARQUIVOS CSVS QUE SERAO TESTADOS
arrayArquivosCSV = [arquivoAtual for arquivoAtual in os.listdir(diretorioCSVs) if arquivoAtual[-4:] == ".csv"]

# CRIANDO UM ARRAY COM O TAMANHO DO JANELAMENTO DE CADA ARQUIVO PRA PLOTAR DEPOIS
arrayJanelamentos = []
for arquivoAtual in arrayArquivosCSV:
    arrayJanelamentos.append(arquivoAtual.split("_")[4][:-2])
```

```python
arrayAcuraciaMediaCadaJanelamento = []
arrayDesvioPadraoCadaJanelamento  = []

# PARA CADA JANELAMENTO (DATASET)
for arquivoCSVAtual in arrayArquivosCSV:
    
    print("Iniciando:", arquivoCSVAtual)
    
    # FACO O KFOLD PARA OBTER A ACURACIA MEDIA E DESVIO
    acuraciaMediaAtual, desvioPadraoAtual = iterarKFold(diretorioCSVs+arquivoCSVAtual, qtdPastas)
    
    # GUARDO O RESULTADO
    arrayAcuraciaMediaCadaJanelamento.append(acuraciaMediaAtual)
    arrayDesvioPadraoCadaJanelamento.append(desvioPadraoAtual)
    
    print("Acurácia Média:", arrayAcuraciaMediaCadaJanelamento[-1])
    print("Desvio Padrão:", arrayDesvioPadraoCadaJanelamento[-1], "\n")
```

### Printando o resultado

```python
for janelamentoAtual, acuraciaMediaAtual, desvioPadraoAtual in zip(arrayJanelamentos, np.array(arrayAcuraciaMediaCadaJanelamento)*100, np.array(arrayDesvioPadraoCadaJanelamento)*100):
    plt.bar(int(janelamentoAtual), acuraciaMediaAtual, width=20, color="#003366")
    plt.errorbar(int(janelamentoAtual), acuraciaMediaAtual, yerr=desvioPadraoAtual, capsize=5, color="#ff0066")
    plt.text(int(janelamentoAtual)-11.75, acuraciaMediaAtual+desvioPadraoAtual+0.1, str(acuraciaMediaAtual)[:4]+"±"+str(desvioPadraoAtual)[:4], fontsize=10)

plt.suptitle("Comparação de Janelamentos")
plt.title("Dataset: SESA v.2 16 kHz 16 bits com remoção de silêncio - Classificador: SGD")
plt.ylabel("Acurácia (%)")
plt.xlabel("Janelamento (ms)")
plt.ylim(min(np.array(arrayAcuraciaMediaCadaJanelamento)*100)-max(np.array(arrayDesvioPadraoCadaJanelamento)*100)-1,max(np.array(arrayAcuraciaMediaCadaJanelamento)*100)+max(np.array(arrayDesvioPadraoCadaJanelamento)*100)+1)
plt.xticks(np.array(arrayJanelamentos).astype("int"))
plt.grid(alpha=0.33, axis="y")
plt.savefig(caminhoSalvar+"AcuraciaXJanelamento.pdf", format='pdf', bbox_inches='tight')
plt.show()
```
