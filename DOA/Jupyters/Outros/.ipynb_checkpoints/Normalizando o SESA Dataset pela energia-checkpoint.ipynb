# Jupyter para normalizar o SESA Dataset

Os arquivos serão abertos, normalizados pela energia e salvos em outra pasta. Para normalizar pela energia, basta dividir pelo desvio padrão. O objetivo disso é facilitar o cálculo da SNR quando for adicionado ruído manualmente.

```python
import matplotlib.pyplot as plt
import librosa
import numpy as np
import os
from IPython.display import Audio
```

```python
def calcularEnergia(sinal):
    return sum(sinal**2)/len(sinal)
```

### Abrindo um áudio qualquer

```python
caminho = "/home/dimi/Downloads/Datasets/SESA/SESA_Original/test/casual_000.wav"
sinalOriginal, freqAmostragem = librosa.load(caminho, sr=None, mono=True)
```

### Normalizando pela amplitude

```python
sinalNormalizadoAmplitude = sinalOriginal/np.amax(sinalOriginal)
```

### Normalizando pela energia

```python
sinalNormalizadoEnergia = sinalOriginal/np.std(sinalOriginal)
```

### Resultados

```python
print("Sinal Original:")
print(calcularEnergia(sinalOriginal))
print(np.var(sinalOriginal), "\n")

print("Sinal Normalizado pela amplitude:")
print(calcularEnergia(sinalNormalizadoAmplitude))
print(np.var(sinalNormalizadoAmplitude), "\n")

print("Sinal Normalizado pela energia:")
print(calcularEnergia(sinalNormalizadoEnergia))
print(np.var(sinalNormalizadoEnergia), "\n")
```

```python
sinalNormalizadoEnergia = sinalOriginal/np.std(sinalOriginal)
plt.plot(sinalNormalizadoEnergia, label="Energia Normalizada")
plt.plot(sinalNormalizadoAmplitude, label="Amplitude Normalizada")
plt.plot(sinalOriginal, label="Original")
plt.legend()
plt.show()
```

```python
print("Sinal Original")
Audio(data=sinalOriginal, rate=freqAmostragem)
```

```python
print("Normalizado pela Amplitude")
Audio(data=sinalNormalizadoAmplitude, rate=freqAmostragem)
```

```python
print("Normalizado pela Energia")
Audio(data=sinalNormalizadoEnergia, rate=freqAmostragem)
```

# Convertendo o SESA

Agora que funcionou, vou aplicar o algoritmo para o SESA Dataset.

#### Pasta de treino

```python
pastaOrigem   = "/home/dimi/Downloads/Datasets/SESA/SESA_Original/train/"
pastaDestino  = "/home/dimi/Downloads/Datasets/SESA/SESA_Normalizado/train/"

# LISTANDO OS ARQUIVOS DA PASTA DE ORIGEM
arrayArquivos = os.listdir(pastaOrigem)
totalArquivos = len(arrayArquivos)

# PARA CADA ARQUIVO EU:
for i, arquivoAtual in enumerate(arrayArquivos):
    
    print("Convertendo arquivo " + str(i+1) + " de " + str(totalArquivos) + " -> " + str(100*((i+1)/totalArquivos)) + "%")
    
    # ABRO 
    sinalOriginalAtual, freqAmostragem = librosa.load(pastaOrigem+arquivoAtual, sr=None, mono=True)

    # NORMALIZO
    sinalNormalizadoAtual = sinalOriginalAtual/np.std(sinalOriginalAtual)
    
    # E SALVO
    librosa.output.write_wav(pastaDestino+arquivoAtual, sinalNormalizadoAtual, freqAmostragem)

print("Operação finalizada")
```

#### Pasta de teste

```python
pastaOrigem   = "/home/dimi/Downloads/Datasets/SESA/SESA_Original/test/"
pastaDestino  = "/home/dimi/Downloads/Datasets/SESA/SESA_Normalizado/test/"

# LISTANDO OS ARQUIVOS DA PASTA DE ORIGEM
arrayArquivos = os.listdir(pastaOrigem)
totalArquivos = len(arrayArquivos)

# PARA CADA ARQUIVO EU:
for i, arquivoAtual in enumerate(arrayArquivos):
    
    print("Convertendo arquivo " + str(i+1) + " de " + str(totalArquivos) + " -> " + str(100*((i+1)/totalArquivos)) + "%")
    
    # ABRO 
    sinalOriginalAtual, freqAmostragem = librosa.load(pastaOrigem+arquivoAtual, sr=None, mono=True)

    # NORMALIZO
    sinalNormalizadoAtual = sinalOriginalAtual/np.std(sinalOriginalAtual)
    
    # E SALVO
    librosa.output.write_wav(pastaDestino+arquivoAtual, sinalNormalizadoAtual, freqAmostragem)

print("Operação finalizada")
```

# Conferindo

```python
caminho = "/home/dimi/Downloads/Datasets/SESA/SESA_Normalizado/train/"

arrayEnergias = []

for arquivoAtual in os.listdir(caminho):
    sinal, freqAmostragem = librosa.load(caminho+arquivoAtual, sr=None, mono=True)
    arrayEnergias.append(calcularEnergia(sinal))
    
print("Média:", np.mean(arrayEnergias))
print("Máximo:", np.amax(arrayEnergias))
print("Mínimo:", np.amin(arrayEnergias))
```
