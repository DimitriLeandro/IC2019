```python
def grausParaRad(angRad):
    return (angRad * np.pi)/180
```

```python
def radParaGraus(angGraus):
    return (angGraus * 180)/np.pi
```

```python
def calcularEnerigaRuidoDadoSNR(snrDesejada, energiaSinal):
    return energiaSinal/(10**(snrDesejada/10))
```

```python
def vetorUnitario(vetor):
    return vetor/np.linalg.norm(vetor)
```

```python
def anguloEntreDoisVetores(v1, v2):
    v1Unitario = vetorUnitario(v1)
    v2Unitario = vetorUnitario(v2)
    return np.arccos(np.clip(np.dot(v1Unitario, v2Unitario), -1.0, 1.0))
```

```python
def anguloAzimutal(vetor):
    return np.arctan(vetor[1]/vetor[0])
```

```python
def anguloElevacao(vetor):
    return anguloEntreDoisVetores([vetor[0], vetor[1], 0], vetor)
```

```python
def calcularVetorWavefront(coordenadasMics, arrayDelays, freqAmostragem, vSom=340):
    A = coordenadasMics[1:]
    b = (np.array(arrayDelays[1:])/freqAmostragem)*vSom
    #return A, b
    w, __, __, __ = np.linalg.lstsq(A, b, rcond=None)
    return w
```

```python
def rmse(a, b):
    return ((a-b)**2)**(1/2)
```

#### Gerando os gráficos 2D

```python
%%time

# ESPECIFICANDO QUAIS SERAO OS ANGULOS VERDADEIROS: None para angulos aleatorios (EM RAD)
azimutalGrausVerdadeiro = randint(0, 90)
elevacaoGrausVerdadeiro = 0
azimutalRadVerdadeiro   = grausParaRad(azimutalGrausVerdadeiro)
elevacaoRadVerdadeiro   = grausParaRad(elevacaoGrausVerdadeiro)

for SNRAtual in range(-30, 31, 10):

    # DEFININDO A ENERGIA DO RUIDO PARA SER DA SNR QUE EU QUERO
    energiaRuido = calcularEnerigaRuidoDadoSNR(snrDesejada=SNRAtual, energiaSinal=1)

    # SIMULANDO UM ARRAY DE MICROFONES COM OS ANGULOS ESPECIFICADOS
    objSimularArrayMics = SimularArrayMicrofones(caminho, energiaRuido, True, azimutalRadVerdadeiro, elevacaoRadVerdadeiro, coordenadasMics)
    arraySinaisSimulados, __, freqAmostragem = objSimularArrayMics.obterResultado()

    # ARRAYS PARA GUARDAR OS RESULTADOS
    arrayAngulosAzimutais     = []
    arrayEnergiasAposDelaySum = []
    
    objBeamforming = GeneralizedSidelobeCanceller()

    # PARA CADA ANGULO AZIMUTAL
    qtd = 100
    for azimutalRadAtual in np.linspace(0, 2*np.pi, qtd):

        # SABENDO OS ANGULOS ATUAIS, PRECISO DEFINIR AS DEFASAGENS DA DIRECAO ATUAL
        arrayDelays = objSimularArrayMics.obterArrayDelays(coordenadasMics, freqAmostragem, azimutalRadAtual, elevacaoRadVerdadeiro)

        # GERANDO O SINAL BEAMFORMADO DE ACORDO COM OS DELAYS ENCONTRADOS
        sinalBeamformado, __ = objBeamforming.gerarSinalBeamformado(arraySinaisSimulados, -np.array(arrayDelays))

        # GUARDANDO AS INFORMACOES NOS ARRAYS DE RESULTADOS
        arrayAngulosAzimutais.append(azimutalRadAtual)
        arrayEnergiasAposDelaySum.append(np.var(sinalBeamformado))
    
    # PRINTANDO
    ax = plt.subplot(111, projection='polar')
    ax.plot(arrayAngulosAzimutais, arrayEnergiasAposDelaySum, label="Energia")
    ax.plot([azimutalRadVerdadeiro, azimutalRadVerdadeiro], [0, np.max(arrayEnergiasAposDelaySum)], label="Azimutal Verdadeiro")
    plt.title("SNR:" + str(SNRAtual) + "dB")
    ax.legend()
    ax.set_rmax(np.max(arrayEnergiasAposDelaySum))    
    ax.grid(alpha=0.5)
    plt.show()
```

#### Varrendo o delay sum por todo o espaço 3D

```python
%%time

# ESPECIFICANDO QUAIS SERAO OS ANGULOS VERDADEIROS: None para angulos aleatorios (EM RAD)
azimutalGrausVerdadeiro = randint(0, 90)
elevacaoGrausVerdadeiro = randint(0, 90)
azimutalRadVerdadeiro   = grausParaRad(azimutalGrausVerdadeiro)
elevacaoRadVerdadeiro   = grausParaRad(elevacaoGrausVerdadeiro)

for SNRAtual in range(-30, 31, 10):
    # DEFININDO A ENERGIA DO RUIDO PARA SER DA SNR QUE EU QUERO
    energiaRuido = calcularEnerigaRuidoDadoSNR(snrDesejada=SNRAtual, energiaSinal=1)

    # SIMULANDO UM ARRAY DE MICROFONES COM OS ANGULOS ESPECIFICADOS
    objSimularArrayMics = SimularArrayMicrofones(caminho, energiaRuido, True, azimutalRadVerdadeiro, elevacaoRadVerdadeiro, coordenadasMics)
    arraySinaisSimulados, __, freqAmostragem = objSimularArrayMics.obterResultado()

    # ARRAYS PARA GUARDAR OS RESULTADOS
    arrayAngulosAzimutais       = []
    arrayAngulosElevacao        = []
    arrayEnergiasAposDelaySum   = []
    
    objBeamforming = GeneralizedSidelobeCanceller()

    # APLICANDO O BEAMFORMING EM TODAS AS DIRECOES DO ESPACO 3D
    passo = 5
    for azimutalAtual in range(0, 180, passo):
        for elevacaoAtual in range(0, 90, passo):

            # SABENDO OS ANGULOS ATUAIS, PRECISO DEFINIR AS DEFASAGENS DA DIRECAO ATUAL
            arrayDelays = objSimularArrayMics.obterArrayDelays(coordenadasMics, freqAmostragem, grausParaRad(azimutalAtual), grausParaRad(elevacaoAtual))

            # GERANDO O SINAL BEAMFORMADO DE ACORDO COM OS DELAYS ENCONTRADOS
            sinalBeamformado, __ = objBeamforming.gerarSinalBeamformado(arraySinaisSimulados, -np.array(arrayDelays))

            # GUARDANDO AS INFORMACOES NOS ARRAYS DE RESULTADOS
            arrayAngulosAzimutais.append(azimutalAtual)
            arrayAngulosElevacao.append(elevacaoAtual)
            arrayEnergiasAposDelaySum.append(np.var(sinalBeamformado))

    # ORGANIZANDO OS ARRAYS
    matrizEnergiasAposDelaySum = np.reshape(arrayEnergiasAposDelaySum, (len(np.unique(arrayAngulosElevacao)), len(np.unique(arrayAngulosAzimutais))))
    angulosAzimutaisUnicos     = np.unique(arrayAngulosAzimutais)
    angulosElevacaoUnicos      = np.unique(arrayAngulosElevacao)

    print("Azimutal Verdadeiro:", azimutalGrausVerdadeiro)
    print("Elevação Verdadeiro:", elevacaoGrausVerdadeiro)
    heatmap = sns.heatmap(matrizEnergiasAposDelaySum, xticklabels=angulosAzimutaisUnicos, yticklabels=angulosElevacaoUnicos, cbar_kws={'label': 'Energia após delay sum'}, cmap=plt.cm.Blues_r)
    plt.title("SNR: " + str(SNRAtual) + " dB")
    plt.xlabel("Ângulo Azimutal")
    plt.ylabel("Ângulo de Elevação")
    plt.show()
```

## 2) Ganho em SNR do beamforming

```python
%%time

passo      = 5
repeticoes = 1

# RANGE DE SNRS
rangeSNRs = np.arange(-30, 31, passo)

# CADA LINHA UM SNR E CADA COLUNA UMA SIMULACAO DIFERENTE
matrizSNRsSaida = []

objBeamforming = GeneralizedSidelobeCanceller()

for snrEntradaAtual in rangeSNRs:
    
    linhaSNRsSaida = []
    
    for i in range(repeticoes):    
        
        # SIMULANDO UM ARRAY DE MICROFONES COM ANGULOS ALEATORIOS E SEM RUIDOOO
        objSimularArrayMics = SimularArrayMicrofones(caminho, 0, True, None, None, coordenadasMics)
        arraySinaisSimuladosPuros, __, freqAmostragem = objSimularArrayMics.obterResultado()
        
        # AQUI EU JA TENHO OS SINAIS SIMULADOS DEFASADOS, CADA UM COM POTENCIA UNITARIA
        # AGORA VOU ADICIONAR O RUIDO
        energiaRuido = calcularEnerigaRuidoDadoSNR(snrEntradaAtual, 1)
        arraySinaisSimuladosRuidosos = objSimularArrayMics.adicionarRuido(arraySinaisSimuladosPuros, energiaRuido)
        
        # GERANDO O SINAL BEAMFORMADO SEM SABER QUAIS SAO OS DELAYS
        arrayDelays = objBeamforming.obterArrayDelays(arraySinaisSimuladosRuidosos)
        sinalBeamformado, __ = objBeamforming.gerarSinalBeamformado(arraySinaisSimuladosRuidosos, arrayDelays)
        
        # ACHANDO O RUIDO E A SNR DE SAIDA
        energiaRuidoSaida = np.var(sinalBeamformado) - np.corrcoef(arraySinaisSimulados[0], sinalBeamformado)[0][1]
        snrSaida          = 10 * np.log10(np.var(sinalBeamformado)/energiaRuidoSaida)
        
        # GUARDANDO AS INFORMACOES NO ARRAYS DE RESULTADOS
        linhaSNRsSaida.append(snrSaida)
        
    matrizSNRsSaida.append(linhaSNRsSaida)
```

```python
arraySNRSaidaMedio = []
desvioSNRSaida     = []

for linha in matrizSNRsSaida:
    arraySNRSaidaMedio.append(np.mean(linha))
    desvioSNRSaida.append(np.std(linha))

plt.errorbar(rangeSNRs, arraySNRSaidaMedio, yerr=desvioSNRSaida)
plt.title("Ganho de SNR após o delay sum")
plt.xlabel("SNR de entrada")
plt.ylabel("SNR de saída")
plt.grid(alpha=0.5)
```

## 3) Diferença entre os ângulos estimados pelo delay sum e os ângulos verdadeiros

```python
%%time

# RANGE DE SNRS PARA TESTAR
rangeSNRs = np.arange(-20, 10, 3)

# CADA LINHA E UMA SNR E CADA COLUNA E UM RMSE PARA CROSS VALIDATION
matrizResultadosAzimutal = []
matrizResultadosElevacao = []

# COMECANDO COM OS PARAMETROS INCICIAS
objBeamforming = GeneralizedSidelobeCanceller()
azimutalGrausVerdadeiro = randint(0, 90)
elevacaoGrausVerdadeiro = randint(0, 90)
azimutalRadVerdadeiro   = grausParaRad(azimutalGrausVerdadeiro)
elevacaoRadVerdadeiro   = grausParaRad(elevacaoGrausVerdadeiro)

for snrAtual in rangeSNRs:
    
    arrayRMSEAzimutal = []
    arrayRMSEElevacao = []
    
    for i in range(10):
    
        # DEFININDO A ENERGIA DO RUIDO PARA SER DA SNR QUE EU QUERO
        energiaRuido = calcularEnerigaRuidoDadoSNR(snrDesejada=snrAtual, energiaSinal=1)

        # SIMULANDO UM ARRAY DE MICROFONES
        objSimularArrayMics     = SimularArrayMicrofones(caminho, energiaRuido, normalizarEnergia=True, azimutalRad=azimutalRadVerdadeiro, elevacaoRad=elevacaoRadVerdadeiro, coordenadasMics=coordenadasMics)
        arraySinaisSimulados, __, freqAmostragem = objSimularArrayMics.obterResultado()

        # CALCULANDO OS DELAYS POR ANALISE DE CORRELACAO   
        objBeamforming = GeneralizedSidelobeCanceller() 
        arrayDelays = objBeamforming.obterArrayDelays(arraySinaisSimulados)
        w = calcularVetorWavefront(coordenadasMics, -np.array(arrayDelays), freqAmostragem)

        # COLOCANDO OS RESULTADOS NOS ARRAYS
        arrayRMSEAzimutal.append((rmse(azimutalGrausVerdadeiro, radParaGraus(anguloAzimutal(w)))))
        arrayRMSEElevacao.append((rmse(elevacaoGrausVerdadeiro, radParaGraus(anguloElevacao(w)))))
        
    # COLOCO A NOVA LINHA CHEIA DE RMSE NA MATRIZ
    matrizResultadosAzimutal.append(arrayRMSEAzimutal)
    matrizResultadosElevacao.append(arrayRMSEElevacao)
```

```python
rmseAzimutalMedioCadaSNR = []
rmseElevacaoMedioCadaSNR = []
desvioAzimutalCadaSNR = []
desvioElevacaoCadaSNR = []

for linha in matrizResultadosAzimutal:
    rmseAzimutalMedioCadaSNR.append(np.mean(linha))
    desvioAzimutalCadaSNR.append(np.std(linha))
for linha in matrizResultadosElevacao:
    rmseElevacaoMedioCadaSNR.append(np.mean(linha))
    desvioElevacaoCadaSNR.append(np.std(linha))

# plt.errorbar(rangeSNRs, rmseAzimutalMedioCadaSNR, yerr=desvioAzimutalCadaSNR, label="Azimutal")
# plt.errorbar(rangeSNRs, rmseElevacaoMedioCadaSNR, yerr=desvioElevacaoCadaSNR, label="Elevação")
plt.plot(rangeSNRs, rmseAzimutalMedioCadaSNR, label="Azimutal")
plt.plot(rangeSNRs, rmseElevacaoMedioCadaSNR, label="Elevação")
plt.legend()
plt.grid(alpha=0.5)
plt.xlabel("SNR [dB]")
plt.ylabel("RMSE [graus]")
```
