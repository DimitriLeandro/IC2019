### Extrao de Features

- Coeficientes Mel-Cepstrais (MFCC)

- Potncia Eficaz (RMS)

- Centroide Espectral

- Largura Espectral

- Contraste Espectral

- Planicidade Espectral

- Rolloff Espectral

- Taxa de Cruzamentos por Zero

```python
melGibson = librosa.feature.mfcc(y=y, sr=sr, n_mfcc=13)
print(melGibson)
```

```python
rms = numpy.mean(librosa.feature.rms(y=y))
print(rms)
```

```python
centroideEspectral = librosa.feature.spectral_centroid(y=y, sr=sr)
print(centroideSpectral)
```

```python
larguraEspectral = librosa.feature.spectral_bandwidth(y=y, sr=sr)
print(larguraEspectral)
```

```python
S = numpy.abs(librosa.stft(y))
contrasteEspectral = librosa.feature.spectral_contrast(S=S, sr=sr)
print(contrasteEspectral)
```

```python
planicidadeEspectral = librosa.feature.spectral_flatness(y=y)
print(planicidadeEspectral)
```

```python
rolloff = librosa.feature.spectral_rolloff(y=y, sr=sr)
print(rolloff)
```

```python
cruzamentosZero = librosa.feature.zero_crossing_rate(y)
print(cruzamentosZero)
```
